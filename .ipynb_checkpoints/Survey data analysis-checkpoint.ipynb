```python
import pandas as pd
```

```python
#Read the first workbook for data in cycle 1

df1 = pd.read_excel(r'Survey data analysis.xlsx', sheet_name = 'Cycle 1', engine='openpyxl', index_col=False)
df1
```

```python
df1.iloc[0,0] = 'Sample Set'
df1.iloc[0,1] = 'ID'
```

```python
#Use the row 0 as the column header

df1.columns = df1.iloc[0]
```

```python
df_c1 = df1.drop([0])
df_c1

#The dataset for cycle 1 has been organized to a proper table for analysis
```

```python
#Column "Sample set" has many missing value; use fillna to fill the values

df_c1['Sample Set'].fillna(method='ffill', inplace=True)
df_c1
```

```python
#Check if there is any cells that are not filled

df_c1.isnull().sum()
```

```python
#Missing data here is entered as "0", need to check where they are.
#The notes on how some data are missing are written in the last column. They have me determine how to deal with the missing data

df_c1[df_c1.values == 0]
```

```python
#Usually in survey research, when a value is completely missing, it means the respondent didn't provide an answer to a question at all. In this case, we can fill the missing value with the row mean.
#For my research purpose and higher analysis accuracy, I won't calculate the mean of the entire row, but the rows that belong to the same question category, and fill the missing value with that mean, since there are only two missing values here.

#First, convert "0" to nan so that "0" won't be included in the calculation of mean

import numpy as np
df_c1=df_c1.replace(0, np.nan)


#Calculate mean for columns C15 to C17 in row 25

rowmean25 = df_c1.iloc[25, [16,17,18]].mean(axis=0) 
rowmean25
```

```python
#Calculate mean for columns A01 to A06 in row 39

rowmean39 = df_c1.iloc[39, [2,3,4,5,6,7]].mean(axis=0)
rowmean39
```

```python
#fill the missing value

values = {'A03': rowmean25, 'C15': rowmean39}
df_c1.fillna(value=values, inplace=True)
```

```python
#check if the missing value has been filled

df_c1.iloc[25, [16,17,18]]
```

```python
#check if the missing value has been filled

df_c1.iloc[39, [2,3,4,5,6,7]]
```

```python
#data for cycle 1 is ready for statistical analysis. 

df_c1
```

```python
#Do the same to clean up data in cycle 2
#Read the second workbook

df2 = pd.read_excel(r'Survey data analysis.xlsx', sheet_name = 'Cycle 2', engine='openpyxl', index_col=False)
df2
```

```python
df2.iloc[0,0] = 'Sample Set'
df2.iloc[0,1] = 'ID'
```

```python
df2
```

```python
df2.columns = df1.iloc[0]
```

```python
df_c2 = df2.drop([0])
df_c2
```

```python
df_c2['Sample Set'].fillna(method='ffill', inplace=True)
df_c2
```

```python
df_c2.isnull().sum()
```

```python
df_c2[df_c2.values == 0]
```

```python
import numpy as np

df_c2 = df_c2.replace(0, np.nan)
```

```python
df_c2.iloc[:,2:19] = df_c2.iloc[:,2:19].astype(float)
```

```python
df_c2['Missing data'].astype(str)
```

```python
#Row 2, 8, and 10 are noted as "Neutral". 
#This is because in our Likert scale questions, there were options fron "strongly agree" to "strongly disagree" but no "neutral", but some respondents indicated "neutral" in the questionnaire. In this case, I will fill the missing data with the value of 2.5.

Neutral = 2.5

for i in range(len(df_c2)):
    if df_c2.iloc[i,19] == 'Neutral':
        df_c2.iloc[2] = df_c2.iloc[i].fillna(Neutral)
    else:
        pass
```

```python
#Row 43 and 83 are noted as "Completely missing". 
#Same as what it was done for cycle 1:
#I won't calculate the mean of the entire row, but the rows that belong to the same question category, and fill the missing value with that mean, since there are only two missing values here.
#Note that this is not the most efficient way to fill the missing data and can cause careless errors. A better way will be writing a for loop as above.

rowmean43 = df_c2.iloc[43, [2,3,4,5,6,7]].mean(axis=0)
rowmean43
```

```python
rowmean83 = df_c2.iloc[81, [8,9,10,12,13,14,15]].mean(axis=0)
rowmean83
```

```python
#fill the missing value

values = {'A05': rowmean43, 'A06': rowmean43, 'B10': rowmean83}
df_c2.fillna(value=values, inplace=True)
```

```python
df_c2
```

```python
#now both df_c1 and df_c2 are ready for t-test
#Next, establish the null and alternative hypothese:
#H0: mean of sample 1 = mean of sample 2
#Ha: mean of sample 1 != mean of sample 2
```

```python
from scipy import stats
```

```python
#for t-test, you need to have the row mean of each ID.

#calculate the row means for df_c1 which will be our sample 1
sample_1 = df_c1.mean(axis=1).to_list()
sample_1
```

```python
#calculate the row means for df_c2 which will be our sample 2
sample_2 = df_c2.mean(axis=1).to_list()
sample_2
```

```python
#calculate the variance of sample_1

var1 = np.var(sample_1)
```

```python
#calculate the variance of sample_2

var2 = np.var(sample_2)
```

```python
#because var1 != var2, set equal_var = False in the t test below:

stats.ttest_ind(sample_1, sample_2, equal_var = False)
```

```python
p = 0.5033688665343059/2
print(p)
```

```python
#p value is 0.25 > 0.05, so we accept the null hypothesis. 
```
